```python
'''
index 객체
- 표 형식의 데이터에서 각 행과 열에 대한 헤더(이름)와 다른 메타데이터(축 이름)을 저장하는 객체
- Series, DataFrame 객체를 생성할 때 사용되는 배열 또는 순차적인 이름이 내부의 색인으로 변환

reindex(재색인)
- 새로운 색인에 맞도록 객체를 새로 생성하는 기능
'''
obj = pd.Series([2.3, 4.3, -4.1, 4.5], index = ['d', 'b', 'a', 'c'])
obj
```

```python
# index 원본을 수정은 지양되는 방식
obj2 = obj.reindex(['a', 'b', 'c', 'd'])
obj2
```

```python
# reindex로 값이 없는(nan) 새 인덱스 추가 가능
obj3 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
obj3
```

```python
# reindex 값 추가
obj4 = obj.reindex(['a', 'b', 'c', 'd', 'e'], fill_value = 0.0)
obj4
```

```python
obj5 = pd.Series(['blue', 'red', 'yellow'], index = [0, 2, 4])
obj5
```

```python
# 전/후 데이터 채우기
obj6 = obj5.reindex(range(6), method = 'ffill')
obj6
```

```python
'''
순위
- 원본 데이터의 순서 정보
'''
s1 = pd.Series([7, -2, 7, 4, 2, 0, 4])
s1
```

```python
s1.rank()  # 중복 랭킹은 평균값으로 반환
```

```python
s1.rank(method = 'first')  # 동점은 먼저 입력된 값을 우선순위
```

```python
s1.rank(ascending = False, method = 'first')  # 내림차순 순위
```

```python
# 동순위처리
print(s1.rank(method = 'max'))     # 동순위 누락
print(s1.rank(method = 'dense'))   # 동순위 동일 처리
```

```python
'''
판다스 기술통계
- 일반적인 수학, 통계 함수 제공
- 처음부터 누락된 데이터를 제외하도록 설계됨
'''
df = pd.DataFrame([
    [1.3, np.nan],
    [7.1, -4.5],
    [np.nan, np.nan],
    [0.75, -1.3]],
    index = list('abcd'), columns = ['one', 'two']
  )
df
```

```python
# 컬럼 합
df.sum()  # nan 제외
```

```python
# 행 합
df.sum(axis = 1)  # 모두 nan일 경우 0 반환
```

```python
# nan 제외 없이 수행
df.sum(skipna = False, axis = 1)
```

```python
# idxmax(), idxmin()    # value가 가장 크거나 작은 값이 있는 인덱스를 리턴 -> 간접 통계
df.idxmax(axis = 1), df.idxmin(axis = 1)
```

```python
# 누산
display(df)
display(df.cumsum())
```
