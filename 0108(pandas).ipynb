```python
# 결측치(누락 데이터) - NaN, null, NA 처리

nan_deck = df['deck'].value_counts(dropna = False)   # nan 값 제외 후 연산 / dropna 매개변수로 처리 지정
print(nan_deck)
print(nan_deck.sum())
```

```python
# 간접적인 방법 : notnull, notna
df.tail().notna()
```

```python
print(df.head().isna().sum())
```

```python
df.isna().sum()
```

```python
'''
데이터 삭제
- 누락 데이터가 있는 컬럼이 데이터 간 상관계수가 낮은 경우
- 누락 데이터가 너무 많은 경우(대형 데이터 셋에서 해당) -> 컬럼 자체를 삭제하는 것이 분석에 유효하다고 판단되는 특수한 경우
'''
```

```python
drop_df = df.dropna(axis = 1)
drop_df.info()
```

```python
# dropna() 수행 시 조건을 부여 : thresh (nan 최소 기준)
drop_df = df.dropna(thresh = 889, axis = 1)    # nan이 889개 초과 시 삭제
drop_df.info()
```

```python
# 특정 컬럼 nan값 삭제(행 삭제)
df_col_loc = df.dropna(subset = ['age'], axis = 0, how = 'any')
len(df_col_loc)
```

```python
'''
누락 데이터 대체
- 기껏 데이터 모아놓고 삭제하지 마라
- 데이터 분석 정확도는 데이터 품질 외 제공되는 데이터 양에 의해서도 상당한 영향을 받음
- 통상 대체값은 데이터 분포의 특성을 가장잘 나타낼 수 있는 값으로 대체(평균, 중앙값, 최빈값)
'''
df.head()
```

```python
df['age'].head(10)
```

```python
cdf = df.copy()
cdf['age'].head(10)
mean_age = cdf['age'].mean()
mean_age

cdf['age'].fillna(mean_age, inplace = True)
cdf['age'].head(10)
```

```python
cdf['age'].isnull().sum()
```

```python
# 최빈값으로 대체(문자열)
cdf['embark_town'][825:830]
mode_embark = cdf['embark_town'].value_counts().idxmax()   # 인덱싱으로 부르면 오류 가능성 조심
cdf['embark_town'].fillna(mode_embark)[825:830]
```

```python
'''
이전값 대체

데이터셋 특성 상 이웃한 데이터는 유사성을 가질 가능성이 높음
'''
cdf = df.copy()
cdf['embark_town'][825:830]
cdf['embark_town'].fillna(method = 'ffill')[825:830]
```
