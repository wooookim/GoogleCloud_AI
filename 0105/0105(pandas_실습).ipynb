```python
# nfl.csv 파일 읽어오기
# - Birthday 열의 값을 날짜/시간으로 변환해서 읽어들이기


```

```python
data['Birthday'] = pd.to_datetime(data['Birthday'])
data['Birthday']
```

```python
# 1) 선수의 이름을 DataFrame의 인덱스로 지정하기


```

```python
df = data.set_index(data['Name']).drop('Name', axis=1)
df
```

```python
# 2-1) 선수의 이름을 DataFrame의 인덱스로 지정하면서
# 2-2) Birthday 열의 값을 날짜 / 시간 형으로 지정하기

df['Birthday'] = pd.to_datetime(df['Birthday'], format='%m/%d/%Y')

df
```
