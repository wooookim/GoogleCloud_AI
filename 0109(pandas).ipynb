```python
import pandas as pd
```

```python
'''
중복 데이터 처리

하나의 데이터셋에 동일한 관측값이 2개 이상 중복되는 경우가 분석 결과를 왜곡 가능성이 있는 경우
-> 중복데이터를 찾아서 제거
'''
df = pd.DataFrame({
    'c1': ['a', 'a', 'b', 'a', 'b'],
    'c2': [1, 1, 1, 2, 2],
    'c3': [1, 1, 2, 2, 2]
})
df
```

```python
# 중복값 존재 여부 반환: duplicated()
# 전체 행에서 각 컬럼 별 중복값이 모두 일치하는 행
# 행 전체가 한 칸씩 내려가면서 일치 여부 확인

df_dup = df.duplicated()
df_dup
```

```python
# 특정 컬럼 별 중복 찾기
col_dup = df['c2'].duplicated()
col_dup
```

```python
# 중복 데이터 제거
df2 = df.drop_duplicates()
df2
```

```python
# 특정 컬럼 중복 데이터 제거
df3 = df['c2'].drop_duplicates()
df3
```

```python
df4 = df.drop_duplicates(subset = 'c2')    # 'c2' 기준으로 중복 제거
df4
```

```python
# 기준 컬럼 다수 지정
df5 = df.drop_duplicates(subset = ['c2', 'c3'])   # 기준 모두 중복이 행만 제거
df5
```

```python
'''
데이터 표준화
- 서로 다른 단위 또는 같은 대상을 다른 형식으로 표현하는 경우
- 분석 정확도 향상을 위한 전처리

 -> 데이터 출처, 데이터 설명이 필요
'''
```

```python
'''
정규화
- 각 컬럼(변수)에 들어있는 숫자 데이터의 상대적크기 차이는 특정 분석 방법에 따라 분석 결과를 달리함
- 수치형 데이터의 상대적 크기 차이를 제거하기 위해 데이터 값을 동일한 크기 기준으로 나눈 비율을 나타내는 것
- 보통 정규화 과정을 거친 데이터 범위는 (0 에서 1 사이) 또는 (-1 부터 1 사이)

1. 해당 열의 최대값(절댓값)으로 나누기
2. 각 열(변수) 데이터 중 최대값과 최소값을 뺀 값으로 나누기

[참고]
- 경우에 따라 특정 분석방법이 상대적 크기 차이를 가정하고 생성되어 있는경우,
  상대적 크기 차이를 갖도록 데이터를 제공하면 분석 결과가 상향될 수 있음
'''
```

```python
df3 = df2.copy()
df3.head()
```

```python
df3['horsepower'] = df3['horsepower'] / abs(df3['horsepower'].max())
df3.head()
```

```python
df4 = df2.copy()
df4.head()
```

```python
# 분자
numerator = df4['horsepower'] - df4['horsepower'].min()
# 분모
denominator = df4['horsepower'].max() - df4['horsepower'].min()

df4['horsepower'] = numerator / denominator
display(df4.head()), df4['horsepower'].describe()
```

```python
'''
범주형 데이터 처리
- 데이터 분석 알고리즘에 따라서는 연속형 데이터보다 일정한 구간으로 나눠서 분석하는게 효율적임
-> 가격, 비용, 효율 등의의미를 가진 연속값을 구간별 차이를 드러내는 방식으로 구현

구간화 binning / pandas -> cut()
'''
```

```python
df2['horsepower'].unique()
```

```python
# 3개 구간으로 분할
# np.histogram(bins =)

count, bin_dividers = np.histogram(df2['horsepower'], bins = 3)
print(count)
print(bin_dividers)
```

```python
# 구간명 지정
bins_names = ['저출력', '보통출력', '고출력']
```

```python
# 각 데이터를 각 구간에 할당
df2['hp_bins'] = pd.cut(x = df2['horsepower'], bins = bin_dividers, labels = bins_names, include_lowest = True)   # include_lowest 첫 경계값 포함 여부
df2[['horsepower', 'hp_bins']].head(10)
```

```python
'''
위 hp_bins(범주형)는 컴퓨터가 인식할 수 없는 형태기 때문에 변경 필요
- 특정 알고리즘에 따라서는 더 높은 숫자값에 가중치를 부여하는 경우가 있음

더미변수
- 숫자 0 또는 1로 표현
- 특성이 있으면 1, 없으면 0
- 원핫인코딩
'''
```

```python
# 범주형 데이터를 더미 변수로 변환
horsepower_dummies = pd.get_dummies(df2['hp_bins'])

horsepower_dummies.head(15)
# ml매서드보다 이게 편함
```

```python
# 날짜 형식의 문자열 리스트
dates = ['2023-01-01', '2023-03-01', '2023-06-01']
```

```python
# object to timestamp
ts_dates = pd.to_datetime(dates)
ts_dates
```

```python
# timestamp to period
# to_period()
# freq 매개변수로 지정

pr_day = ts_dates.to_period(freq = 'D')   # 일 단위로 지정

pr_month = ts_dates.to_period(freq = 'M')   # 월 단위로 지정

pr_year = ts_dates.to_period(freq = 'Y')   # 연 단위로 지정
```

```python
# 시계열 데이터 만들기
# date_range()
ts_data = pd.date_range(
    start = '2023-01-01',  # 날짜 범위시작
    end = None,            # 날짜 범위 끝
    periods = 6,           # 생성할 데이터 수
    freq = 'MS',           # 시간간격(MS: 월의 시작일)
    tz = 'Asia/Seoul')     # 시간대
ts_data
```

```python
# 기간 단위 날짜 데이터 배열
# period_range()

# 1개월 단위
pr_data = pd.period_range(
    start = '2023-01-01',  # 날짜 범위시작
    end = None,            # 날짜 범위 끝
    periods = 6,           # 생성할 데이터 수
    freq = 'M',           # 시간간격(MS: 월의 시작일)
    )
pr_data
```

```python
# 분기
pr_3m_data = pd.period_range(
    start = '2023-01-01',  # 날짜 범위시작
    end = None,            # 날짜 범위 끝
    periods = 6,           # 생성할 데이터 수
    freq = '3M',           # 시간간격(MS: 월의 시작일)
    )
pr_3m_data
```
